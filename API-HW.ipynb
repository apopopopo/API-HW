```python
# import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import urllib
%matplotlib inline

# import API call dependencies
import requests 
from citipy import citipy
import openweathermapy as ow
```

# Generate Cities list

```python

lats = np.random.uniform(low=-90, high=90, size=1500)
lngs = np.random.uniform(low=-180, high=180, size=1500)
```

```python
# put latitude and longitude together
lat_lngs = zip(lats, lngs)
cities_list = []

for lat_lng in lat_lngs:
    # find city closest of lat and long values
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # add non-repeat cities
    if city not in cities_list:
        cities_list.append(city)
```

```python
# check how many cities we got
len(cities_list)
```

# Perform API Calls

```python
# import my API key
api_key = "414647d004dc123309353ec7aaba400f"

# base url
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID="+api_key



city_weather_data=[]
print("Beginning Data Retrieval")
print("------------------------")
for city in cities_list:
    
    try:
        
        city_url=url+"&q="+urllib.request.pathname2url(city)
        
        print(f"Processing record for {city}")
        print(city_url)
        
    
        city_weather=requests.get(city_url).json()
        city_weather_data.append(city_weather)
        
    except:
        print('error..skipping')
        
print("------------------------")
print("Data Retrieval Complete")
print("------------------------")
```

```python
print(city_weather_data)
```

```python
#city weather data df
city_weather_data_df = pd.DataFrame(city_weather_data)
#city weather data csv
city_weather_data_df.to_csv("city_weather_data.csv", sep =",")
```

```python
city_name=[]
cloudiness=[]
country=[]
date=[]
humidity=[]
lat=[]
lng=[]
max_temp=[]
wind_speed=[]

for i in range(len(city_weather_data)):
    try:
        city_name.append(city_weather_data[i]["name"])
        cloudiness.append(city_weather_data[i]["clouds"]["all"])
        country.append(city_weather_data[i]["sys"]["country"])
        date.append(city_weather_data[i]["dt"])
        humidity.append(city_weather_data[i]["main"]["humidity"])
        lat.append(city_weather_data[i]["coord"]["lat"])
        lng.append(city_weather_data[i]["coord"]["lon"])
        max_temp.append(city_weather_data[i]["main"]["temp_max"])
        wind_speed.append(city_weather_data[i]["wind"]["speed"])
    except:
        pass

    
```

```python
city_weather_data_summary={"City":city_name,
                           "Cloudiness":cloudiness,
                           "Country":country,
                           "Date(in Unix Time)":date,
                           "% Humidity":humidity,
                           "Latitude":lat,
                           "Longitude":lng,
                           "Max Temp (F)":max_temp,
                           "Wind Speed(mph)":wind_speed}
city_weather_data_summary_df= pd.DataFrame(city_weather_data_summary)

city_weather_data_summary_df =city_weather_data_summary_df[["City",
                                                  "Cloudiness",
                                                  "Country",
                                                  "Date(in Unix Time)",
                                                  "% Humidity",
                                                  "Latitude",
                                                  "Longitude",
                                                  "Max Temp (F)",
                                                  "Wind Speed(mph)"]]
city_weather_data_summary_df.head()
```

# Latitude vs Temperature Plot

```python
latitude = city_weather_data_summary_df["Latitude"]
max_temp = city_weather_data_summary_df["Max Temp (F)"]
plt.scatter(latitude,max_temp)
plt.grid(True)
plt.title("City Latitude vs. Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Max Temp (F)")
plt.savefig("temp_v_lat.png")
```

# Latitude vs. Cloudiness Plot

```python
cloud = city_weather_data_summary_df["Cloudiness"]
latitude = city_weather_data_summary_df["Latitude"]
plt.scatter(latitude,cloud)
plt.grid(True)
plt.title("City Latitude vs. Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness")
plt.savefig("cloud_v_lat.png")
```

# Latitude vs. Humidity Plot

```python
humidity= city_weather_data_summary_df["% Humidity"]
latitude = city_weather_data_summary_df["Latitude"]
plt.scatter(latitude,humidity)
plt.grid(True)
plt.title("City Latitude vs. % Humidity")
plt.xlabel(" Latitude")
plt.ylabel("Humidity (%)")
plt.savefig("hum_v_lat.png")
```

# Latitude vs. Wind Speed Plot

```python
wind_speed = city_weather_data_summary_df["Wind Speed(mph)"]
latitude = city_weather_data_summary_df["Latitude"]
plt.scatter(latitude,wind_speed )
plt.grid(True)
plt.title("City Latitude vs.Wind Speed(mph)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed(mph)")
plt.savefig("wind_v_lat.png")
```
